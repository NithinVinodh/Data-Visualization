```python
nationality_counts = data['nationality'].value_counts()
specified_countries = ['India', 'England', 'South Africa', 'West Indies', 'Australia', 'New Zealand', 'Sri Lanka']
specified_countries_counts = nationality_counts[nationality_counts.index.isin(specified_countries)]
others_count = nationality_counts[~nationality_counts.index.isin(specified_countries)].sum()
specified_countries_counts['Others'] = others_count
plt.figure(figsize=(8, 8))
plt.pie(specified_countries_counts, labels=specified_countries_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Nationality Distribution of Players in IPL Auction 2023')
plt.axis('equal')
plt.show()
```
